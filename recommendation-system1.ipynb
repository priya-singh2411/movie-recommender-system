```python
import numpy as np
import pandas as pd
credit = pd.read_csv("tmdb_5000_credits.csv")
movie_data = pd.read_csv("tmdb_5000_movies.csv")
```

```python
credit_movie_id = credit.rename(index=str,columns={"movie_id":"id"})
movies_data_merge = movie_data.merge(credit_movie_id,on= 'id')
```

```python
movie_cleaned_df = movies_data_merge.drop(columns=['homepage','title_x','title_y','status','production_companies'])
movie_cleaned_df.head()
```

# Using Weighted average for each movie's average rating.
#This method is used by IMDB to compute the movie rating
#W = (Rv+Cm)/(v+m)
#W = weighted rating 
#R = average for the movie as a number from 0 to 10(mean) = rating
#v = no of votes for the movie
#m = minimum votes required to be listed in the top 250 (currently 3000)
#c = mean vote accross the whole report.

```python
#v= movie_cleaned_df['vote_count']
#R = movie_cleaned_df['vote_average']
#c = movie_cleaned_df['vote_count'].mean()
#m = movie_cleaned_df['vote_count'].quantile(0.70)   #it will take 70 percentile and above value.

v=movie_cleaned_df['vote_count']
R=movie_cleaned_df['vote_average']
C=movie_cleaned_df['vote_average'].mean()
m=movie_cleaned_df['vote_count'].quantile(0.70)       #it will take 70 percentile and above value.
```

```python
#movie_cleaned_df['weighted_average']= ((R*v)+(c*m))/(v+m)
#movie_cleaned_df.head()
movie_cleaned_df['weighted_average']=((R*v)+ (C*m))/(v+m)
movie_cleaned_df.head()
```

```python
movie_cleaned_df_sorted = movie_cleaned_df.sort_values('weighted_average',ascending=False)
movie_cleaned_df_sorted[['original_title','vote_count','vote_average','weighted_average','popularity']].head(20)
```

```python
import matplotlib.pyplot as mp
import seaborn as sb
weighted_avg = movie_cleaned_df_sorted.sort_values('weighted_average', ascending=False)
mp.figure(figsize=(12,6))
axis1 = sb.barplot(x=weighted_avg['weighted_average'].head(10),y=weighted_avg['original_title'].head(10) )
mp.xlim(4,10)
mp.title('BEST MOVIES BY AVERAGE VOTES', weight='bold')
mp.xlabel('Weighted average score',weight='bold')
mp.ylabel('Movie title',weight='bold')
mp.savefig('BEST_MOVIES_BY_AVERAGE_VOTES.png')
```

```python
popularity = movie_cleaned_df_sorted.sort_values('popularity',ascending=False)
mp.figure(figsize=(12,6))
axis1 = sb.barplot(x=popularity['popularity'].head(10),y=popularity['original_title'].head(10))
#mp.xlim(4,10)
mp.xlabel('Movie Popularity',weight='bold')
mp.title('BEST MOVIES BY POPULARITY',weight='bold')
mp.ylabel('Movie Title',weight='bold')
mp.savefig('BEST_MOVIES_BY_POPULARITY.png')
```

# Recommendation based on scaled weighted average and popularity score(Priority is given 50% to both)

```python
from sklearn.preprocessing import MinMaxScaler
scaling = MinMaxScaler()
movie_scaled_df =  scaling.fit_transform(movie_cleaned_df[['weighted_average','popularity']])
movie_normalized_df = pd.DataFrame(movie_scaled_df,columns=['weighted_average','popularity'])
movie_normalized_df.head()
```

```python
movie_cleaned_df[['normalized_weighted_average','normalized_popularity']] = movie_normalized_df
movie_cleaned_df.head()
```

```python
movie_cleaned_df['score'] = movie_cleaned_df['normalized_weighted_average']*0.5 + movie_cleaned_df['normalized_popularity']*0.5
movie_score_df = movie_cleaned_df.sort_values('score',ascending=False)
movie_score_df[['original_title','normalized_weighted_average','normalized_popularity','score']].head()
```

```python
scored_df = movie_cleaned_df.sort_values('score',ascending=False)
mp.figure(figsize=(16,6))
ax = sb.barplot(x=scored_df['score'].head(10),y=scored_df['original_title'].head(10))
mp.xlabel('Score',weight='bold')
mp.ylabel('Movie Title',weight='bold')
mp.title('Best rated and most popular movie blend',weight='bold')
```
